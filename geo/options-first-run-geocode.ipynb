```python
def get_megye(cd):
    megyek={'HR':'Hargita',
    'CV':'Kovászna',
    'MS':'Maros',
    'BV':'Brassó',
    'NT':'Neamț'}
    if cd in megyek:return megyek[cd]
    else:return 'Más megye'
m_geo={}
for i in hun_geo:
    megye=get_megye(hun_geo[i]['megye'])
    if megye not in m_geo:m_geo[megye]=[]
    m_geo[megye].append(str(i).strip())
for i in m_geo:
    m_geo[i]=np.sort(m_geo[i])
```
